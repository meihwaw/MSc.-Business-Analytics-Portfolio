```python
"""
DocString
A1 Assignment Text-Adventure Game
Game: Top Gun 
Author: Mei Hwa Wong

This game is based on the movie Top Gun that tells a story of a young and 
talented pilot selected to join the elite flying school to learn aerial combat.
The player in the game will go through the journey as a pilot in Top Gun
and the goal is to win the Top Gun Trophy.

Round 1: Quiz about the types of planes used in Top Gun academy
Round 2: Decision making during dog fight
Round 3: Decision making during final mission

"""
# importing packages for exit function
from sys import exit

# to upload url of Top Gun opening scene song from You Tube
url ="https://www.youtube.com/watch?v=NI8U9Bc4XOE&list=OLAK5uy_kCsa_Y8wEEkgW5bjmzq6VMY8EcAbCAGrM"
print(f"""
    Welcome to Top Gun!
    For a more immersive experience, you may play this YouTube video in the background:\n{url}
      """)

# importing package to create animation in text
import sys, time, os

# defined function for typewriter animation in text
def typewriter(text):
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(0.03)

        
# defined function to start the game that describes the game from the Top Gun movie
def start():
    
    # prompt to get user to press enter to continue
    input("<Key in any characters to continue> ")
    
    # introducting the game 
    game_introduction = """
    You are one of the top 1% of pilots in the Navy and have been selected
    to attend the elite school where you will learn aerial combat and become
    the best fighter pilots in the world.
    
    The official name of the school is called Fighter Weapons School. The flyers
    call it:
    
    """
    # typewriter animation for game introduction
    typewriter(game_introduction)
    
    # top gun ascii art
    top_gun_ascii_art = """
    
     ▄███████████████████████▄█████▐█████▄   ▄█████▄██ ██ ███ ▐██████████████████▄
         ═▄▄▄▄▄▄▄▄▄▄▄▄▄▄▄▄▄▄ ██▌ ▐██ ▐██▐██  ██▌ ██▌ ██▌███ ██ ████▐██ ███████████████
           ▀▀▀▀▀▀▀▀▀▀▀▀▀▀▀▀▀ ██▌ ▐██ ▐██▐██▄▄██▌ ██▌▄▄▄▄███ ██ ███████ ▀▀▀▀▀▀▀▀▀▀▀▀▀▀
              ▀█████████████ ██▌ ▐██ ▐██▐███▀▀   ██▌▀██▌███ ██ ███▀███ ██████████▀
                  ▄▄▄▄▄▄▄▄▄▄ ██▌ ▐██▄███▐██      ███▄██▌███▄██ ███ ███ ▄▄▄▄▄▄▄
                    ▀▀▀▀▀▀▀▀ ▀▀▀   ▀▀▀▀ ▀▀▀       ▀▀▀▀   ▀▀▀▀▀ ▀▀▀ ▀▀▀ ▀▀▀▀▀
    """
    
    # to print ascii art of the Top Gun logo
    print(top_gun_ascii_art)
    
    # input to prompt user to press enter to continue
    input("Key in any characters to continue\n ")
    
    # to call the new_game function
    new_game()

# defining the start of the entire game
def new_game():
    
    # input to get user to insert their call sign for the game
    call_sign = input("What is your call sign? Definition: A call sign is a nickname in the navy ").title()
    
    # print the introduction of the game
    print(f"""
    Hello {call_sign}, and welcome to Top Gun! To win the Top Gun 
    trophy, you are required to successfully pass 3 levels in the game.
    Your performance will be measured according to your knowledge and how you
    react to different situations. Good luck!
    """)
    
    # input to prompt user to press continue to continue the game 
    input("<Key in any characters to continue>\n ")
    
    # to call out the plane_type function that is the first level of the game
    plane_type()    

# defining the first level of the game that is to identify the type of planes used in Top Gun training
def plane_type():
    
    # print the question for user to answer in the plane_type stage
    print("""
    What are the type of planes used during trainings?
    
    a. aeroplanes
    b. fighter jets
    c. stunt jets
    """)
    
    # prompt user to input answer
    answer_1 = input("Please input one answer: a, b or c ").lower()
    
    # allowable attempt for this round is the initial round and 2 rounds after invalid attempt
    attempt = 1
    
    # while function to ensure user inputs the right answer before moving on to the next stage
    while answer_1 not in ['b'] and attempt > 0:
       
        # to keep count of the attempts user have left
        attempt -= 1
        
        # to print invalid input and notify user how many attempts user have 
        answer_1 = input(f"Invalid input! Please try again! You have {attempt}\
 attempts left. ").lower()
    
    # if statement if user inputs the correct answer and moves to the next level of the game
    if answer_1 == 'b' and len(answer_1) == 1:
        
        # prints the statement if user chooses option
        print("""
        Awesome! That is correct! Congrats on making it to this stage! We will
        now test your agility.
         """)
        
        # input to prompt user to press enter to continue
        input("<Key in any characters to continue> ")
        
        # directs user to option_stage which is the next level of the game
        option_stage()
         
    # user gets directed to new_attempt to ask if user wants to play the plane_type stage again
    else:
        
        # to print to notify user that user have failed in this stage
        print("\nYou have failed in this stage! Please try again ")
        
        # direct user to new attempt to ask whether user wants to continue the game or exit
        new_attempt()

# defining user function for new_attempt option if user has failed the plane_type stage
def new_attempt():
    
    # to obtain input from user whether user wants to replay the first stage - plane_type
    new_attempt = input("Replay this stage? y / n ")
     
    # while function to ensure user input only answers y or n 
    while new_attempt != "n" and new_attempt != "y":
        
        # to prompt input again if user does not enter the correct inputs
        new_attempt = input("Invalid input! Please try again!\nReplay? y / n ")
    
    # direct user to the plane_type stage if user answers y in the input
    if new_attempt == "y":
        
        # direct user to plane_type stage if user selects y which is to replay this stage
        plane_type()
     
    # exit user from the game if user chooses n in the input
    elif new_attempt == "n":
        
        # print statement to user if user chooses n which is to exit the game
        print("Thank you for your time at Top Gun. Goodbye! ")
        
        # exits user from the game 
        exit(0)

# defining the function to for the option_stage that comes after user successfully complete plane_type stage
def option_stage():
    
    # to print description of option_stage
    print("Which task would you like to do? ")
    
    # to print the description of the option_stage for users to pick
    print("""
    a. Let's go for a dogfight! Dogfight is an aeriel battle between fighter aircraft conducted at close range.
    b. I would like to improve my training skills first.
    c. I'm not fit! I want to quit!
          """)
    
    # while loop to ensure user selects the options that are available
    while True:
        
        # input prompt to get user to input choice for this stage ensuring input will be lowercase
        options = input("Please input your answer: a, b or c ").lower()
    
        # if user chooses option a, the while loop breaks and directs user to dog_fight stage
        if "a" in options and len(options) == 1:
            
            # print statement to show output for selecting option a
            print("\nGet your wingman and let's go for a dogfight! ")
            
            # direct user to dog fight stage
            dog_fight()
            
            # breaks the while loop if user chooses this option
            break
        
        # if user chooses option b, the while loop breaks and directs user back to option_stage
        elif "b" in options and len(options) == 1:
            
            # prints statement to show output for selecting option b
            print("\nYou've done enough training now. What would you like to do? ")
            
            # directs user back to the start of option stage
            option_stage()

            # breaks the while loop if user picks this option
            break
        
        # if user chooses option c, the while loop breaks and exits the game... No new attempt is given here
        elif "c" in options and len(options) == 1:
            
            # prints statement to show output if user chooses option c
            print("""You have made your choice. I hope you have a great journey 
                  ahead. Goodbye
                  """)    
            
            # directs user to exit game
            exit()
            
            # breaks the while loop if user picks this option
            break
        
        # prints statement if user does not input any of the statements listed in if and elif
        else:
            
            # prints output if user enters invalid input
            print("Invalid input, please try again! ")

# definining the dog_fight stage that comes when user select optino a in option_stage
def dog_fight():
    
    # to ask for input on wingman's name 
    wingman = input("What is your wingman's name? ").title()
    
    # prints the description of the dog_fight stage and user will have to pick their option
    print(f"""
    You and your wingman, {wingman} are competing against Iceman and Hollywood
    in a dogfight but at the same time has a mission to capture a MiG-28s; a supersonic
    fast fighter aircraft. 

    You spotted the MiG-28s while still in the middle of the dogfight. Do you proceed to
    capture the MiG-28s and abandon Goose or continue to stay in the dogfight?

    Pick your option:    
    a. capture the MiG-28s
    b. stay in the dogfight
    
          """)
    
    # user to input their decision in the dog_fight stage
    decisions = input("Please input one answer: a or b ").lower()
    
    while True:
    
        # if statement if user chooses option a and input capture, the while loop breaks and user fails the game
        if "a" in decisions and len(decisions) == 1:
            
            # print statement to show output for selecting option a
            print("\nAbandoning your teammate is poor demonstration of teamwork! ")
            
            # directs user to the fail function to indicate user have failed the game
            fail()
            
            # breaks the loop if user picks this option
            break
     
        # elif statement if user chooses option b which is the correct answer and directs user to the final mission
        elif "b" in decisions and len(decisions) == 1:
            
            # print statement to show output for selecting option b
            print("""
            Awesome! You understood the importance of teamwork in naval aviation.
            Only teamwork can make the dream work!
            """)
            
            # directs user to direct mission function to continue the game
            final_mission()
    
        # prints invalid input if user does not input any of the option or keywords in the decisions input...
        # ... it will direct user back to the start of the dog_fight stage
        else:
            
            # prints the statement to show the output of not inputing a valid function
            print("Invalid input! Please try again. ")
            
            # directs user to the fail function and user fails the game and will have to replay from the start
            fail()

# definining the final mission of the game which is also the last stage of the game
def final_mission():
    
    # prints the description of the final stage
    print("""Great job soldier! You've been selected for the final mission!
    
    You have received immediate deployment to deal with a crisis situation.
    You are sent with your team to Enterprise to provide air support to the
    rescue of the SS Payton, a disabled communication ship that drifted into the
    waters.
    
    While lowering from the altitude to approach the ship, your radar
    screen showed there are 6 MiGs targeting you. What would you do?
    
    a. attack the MiGs together with your teammates
    b. run away to base
    
    """)
    
    # while loop statement to control user input to only be limited to the options available
    while True:
        
        # to prompt user input for the action user will choose for this stage
        action = input("Please input one answer: a or b ").lower()
    
        # if statement and while loop breaks if user inputs the correct input and directs users to victory
        if "a" in action and len(action) == 1:
            
            # prints the statement to show the output of selecting option a
            print("""Great fighter skill soldier! You've managed to bring down
                 the enemy's MiGs! Great save!
                 """)
            
            # directs user to winner function
            winner()
            
            # breaks the loop if user selects option a
            break
        
        # if statement and while loop breaks if user inputs the correct input and direct users to fail function
        elif "b" in action and len(action) == 1:
            
            # print statement to show the output for selecting optino b
            print("You have expose our base to the enemy! We've been attacked! ")
            
            # directs user to the fail function to indicate user have failed this stage and lost the game
            fail()
            
            # break the loop if user selects optino b
            break
        
        # if inputs are invalid, print invalid input statement and direct user back to the start of final mission
        else:
            
            # print statement if user enters invalid input
            print("Invalid input! Please try again. ")
            
            # direct user to final mission function
            final_mission()

# defining the winner function
def winner():
    
    # print statement if user won the game
    print("Congratulations! You've won the Top Gun Trophy for your exceptional skills and attitude! ")
    exit(0)

# defining the winner function
def fail():
    
    # print statement if user lost the game
    print("\nYou have failed in your mission. Game over! ")
    
    # prompt play again input to ask if user wants to play again
    play_again = input("\nDo you want to play again? y / n ").lower()
    
    # while loop to control user input to either be y or n
    while play_again != "y" and play_again != "n":
        
        # print statement if user inputs statement other than y or n
        print("\n Invalid input. Please enter y or n ")
    
    # if statement for when user chooses y
    if play_again == "y":
        
        # directs user to new_game stage if user selects y
        new_game()        
    
    # elif statement for when user chooses n
    elif play_again == "n":
        
        # print statement if user selects n
        print("\nThank you for your time at Top Gun. Goodbye!")      
        
        # exits user from game
        exit(0)

# Starting the game
start()
```
