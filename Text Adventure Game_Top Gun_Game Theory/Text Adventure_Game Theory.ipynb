```python
# A2 Assignment: Game Theory - The Python Equilibrium
# Author: Mei Hwa Wong

# import random package
import random

# import math package
import math

# import table package
from tabulate import tabulate

# defining game theory as one whole game block
def new_game():
    
    # prompt user to play a Game Theory simulation whether to play against Python or have two Python players play yes: to play with Python no: to have 2 Python players play
    user_play = input(prompt = """Hello, would you like to play a game theory simulation with Python? Select yes or no """).lower()
    print("\n" + user_play)
    
    # to ensure that user is inputting the valid response that is only yes or no
    while (not user_play.startswith("yes") or len(user_play)>3) and (not user_play.startswith("no") or len(user_play)>2):
        user_play = input("Invalid input, please answer yes or no").lower()
        
    # conditional game route if user wants to play against Python
    if "yes" in user_play:
        user_name = input("Hello and welcome to Adventure Coffee! You will play against me, Python! What is your name?").title()
        
        # game introduction to explain to users the background of the game and how to win
        print(f"""
        Hi {user_name}, you and I are both the best coffee shop owners in Coffeeland!
        There is a new type of coffee bean in town. Do not switch! We are not sure if this will impact the coffee quality!
        I'm not sure about you but I don't think we should change it.
        
        Upon returning to your coffee shop, you are curios and interested in this new vendor.
        Afterall, this is your business and what you decide for your business is private and confidential.
        
        Player with the highest points wins. In this game, the points you gain is the profit you made from your coffee shop!
        """)
        
        # game theory matrix showing the payoff for players depending on the strategy selected by the opponent
        table = [["Change", "Both gain 5", f"{user_name}; 10, Python; 0"],
                ["No change", f"{user_name}; 0, Python; 10", "Both gain 1"]]
        
        # to print the table showing the payoff for each player depending on their strategy
        print("\n" + "\033[1m" + f"{user_name} vs Python" + "\033[0m")
        print(tabulate(table, headers=["Change", "No change"], tablefmt = "fancy_grid", numalign = "center"))
        
        # pay-off matrix
        matrix = [[0,5],[1,10]]
        
        # score board dictionary C = change, NC = no change
        dictionary_scores = {"(C, C)": (matrix[0][1], matrix[0][1]),
                             "(NC, C)": (matrix[0][0], matrix[1][1]),
                             "(C, NC)": (matrix[1][1], matrix[0][0]),
                              "(NC, NC)": (matrix[1][0], matrix[1][0])}
        
        # prompt user to provide a strategy response and ensure only response to the valid response (C or NC)
        while True:
            user_strategy = input("""
            Do you want to change or do not want to change the coffee beans?
            Select:
            C as change
            NC as no chance
            """).upper()
            if user_strategy in ["C", "NC"]:
                break
            else:
                print("Invalid input, please select 'C' or 'NC'")
                
        # Python player chooses strategy by random; 0: change, 1: no change
        python_strategy = random.randint(0,1)
        
        # conditional structure with outputs of each scenario
        if user_strategy == "C" and python_strategy == 1:
            user_score = dictionary_scores["(C, NC)"][0]
            python_score = dictionary_scores["(C, NC)"][1]
            print(f"""You changed to the new coffee beans & I did not! Customers prefers your coffee now!
            {user_name} receive {user_score} points
            Python receive {python_score} points
            """)
            
        elif user_strategy == "C" and python_strategy == 0:
            user_score = dictionary_scores["(C, C)"][0]
            python_score = dictionary_scores["(C, C)"][1]
            print(f"""We both changed the coffee beans! The new beans are certainly more flavorful
            and fragrant! Both our business are going to flourish!
            {user_name} receive {user_score} points
            Python receive {python_score} points 
            """)
            
            
        elif user_strategy == "NC" and python_strategy == 0:
            user_score = dictionary_scores["(NC, C)"][0]
            python_score = dictionary_scores["(NC, C)"][1]
            print(f"""You did not change! But I did! The new coffee is certainly better!
            {user_name} receive {user_score} points
            Python receive {python_score} points
            """)
            
        elif user_strategy == "NC" and python_strategy == 1:
            user_score = dictionary_scores["(NC, NC)"][0]
            python_score = dictionary_scores["(NC, NC)"][1]
            print(f"""We both did not change the coffee beans! But it seemed like the
            new coffee beans are definitely better!
            We missed out on an opportunity!
            {user_name} receive {user_score} points
            Python receives {python_score} points
            """)
            
        else:
            print("Something is wrong")
            
        # declare the winner
        if user_score > python_score:
            print("""Congratulations you won!""")
            
        elif user_score == python_score:
            print("""We tie!""")
            
        elif user_score < python_score:
            print("""Congratulations to Python!""")
            
        # to prompt if user wants to replay the game from the beginning
        replay = input("Do you want to start again? Select yes or no ").lower()
        while (replay != "yes" or len(replay) != 3) and (replay != "no" or len(replay) != 2):
            replay = input("Invalid input, please select yes or no ").lower()       
        if replay == "yes":
                new_game()         
        else:
            print("\nGoodbye!")
            
    # game route condition if user does not want to play and have 2 Python players play against each other
    if "no" in user_play:
        print("""Let's watch 2 Python players play against each other!""")
            
        # introducting python players
        print(f"""We will have Python1 and Python2 as our players""")
            
        # game theory matrix showing the payoff for players depending on the strategy selected by the opponent
        table = [["Change", "Both gain 5", "Python1; 10, Python2; 0"],
                ["No change", "Python1; 0, Python2; 10", "Both gain 1"]]
        
        # to print the table showing the payoff for each player depending on their strategy
        print("\n" + "\033[1m" + "Python1 vs Python2" + "\033[0m")
        print(tabulate(table, headers=["Change", "No change"], tablefmt = "fancy_grid", numalign = "center"))
        
            
        # declare variable for python1 and python2 scores 
        python1_total_scores = 0
        python2_total_scores = 0
            
        # to create loop to so that both python players are playing for 5 rounds
        for game in range(5): 
        # from matrix
            print(f"\nRound {game+1}")
                
         # pay-off matrix
            matrix = [[0, 5],[1,10]]
                
        # score board dictionary C = change, NC = no change
            dictionary_scores = {"(C, C)": (matrix[0][1], matrix[0][1]),
                                "(NC, C)": (matrix[0][0], matrix[1][1]),
                                "(C, NC)": (matrix[1][1], matrix[0][0]),
                                "(NC, NC)": (matrix[1][0], matrix[1][0])}
                
            # Python1 chooses strategy by random - 0: change or 1: no change
            python1_strategy = random.randint(0,1)
                
            # Python2 chooses strategy by random - 0: change or 1: no change
            python2_strategy = random.randint(0,1)
                
            # conditional statement of different scenario outputs
            if python1_strategy == 0 and python2_strategy == 0:
                python1_score = dictionary_scores["(C, C)"][0]
                python2_score = dictionary_scores["(C, C)"][1]
                python1_total_scores += python1_score
                python2_total_scores += python2_score
                print(f"""\nWe both changed the coffee beans! The new beans are certainly more flavorful
                and fragrant! Both our business are going to flourish! 
                    
                Python1 receives {python1_score} points
                Python2 receives {python2_score} points
                """)
                    
            elif python1_strategy == 1 and python2_strategy == 0:
                python1_score = dictionary_scores["(NC, C)"][0]
                python2_score = dictionary_scores["(NC, C)"][1]
                python1_total_scores += python1_score
                python2_total_scores += python2_score
                print(f"""\nPython1 did not change! But Python2 did! The new coffee is certainly better!
                    
                Python1 receives {python1_score} points
                Python2 receives {python2_score} points
                """)
                    
            elif python1_strategy == 0 and python2_strategy == 1:
                python1_score = dictionary_scores["(C, NC)"][0]
                python2_score = dictionary_scores["(C, NC)"][1]
                python1_total_scores += python1_score
                python2_total_scores += python2_score
                print(f"""\nPython1 changed the coffee beans, Python2 did not! Customers prefer Python1's coffee!
                    
                Python1 receives {python1_score} points
                Python2 receives {python2_score} points
                """)
                    
            elif python1_strategy == 1 and python2_strategy == 1:
                python1_score = dictionary_scores["(NC, NC)"][0]
                python2_score = dictionary_scores["(NC, NC)"][1]
                python1_total_scores += python1_score
                python2_total_scores += python2_score
                print(f"""\nWe both did not change the coffee beans! Seems like they are a better quality.
                We missed out on an opportunity!
                    
                Python1 receives {python1_score} points
                Python2 receives {python2_score} points
                """)
                    
            else:
                print("Something is wrong")
         
        # to print the total scores for each python players accumulated from all 5 rounds
        print("Python1 total score:", python1_total_scores)
        print("Python2 total score:", python2_total_scores)
                
        # Declare winner   
        if python1_total_scores > python2_total_scores:
            print("\nCongratulations Python1, you win!")
                    
        elif python1_total_scores == python2_total_scores:
            print("\nIt's a tie!")
                    
        else:
            print("\nCongratulations Python2, you win!")
        
        # to prompt user if they want to replay the game from the beginning
        replay = input("Do you want to start again? Select yes or no ").lower()
        while (replay != "yes" or len(replay) != 3) and (replay != "no" or len(replay) != 2):
            replay = input("Invalid input, please select yes or no").lower()
        if replay == "yes":
            new_game()    
        else:
            print("Goodbye!")

# to begin the game                        
new_game()
```
